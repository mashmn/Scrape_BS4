```python
f=open("Power_Rangers.txt",mode='r',encoding='utf-8')
#f.readline()
```

## Beautiful Soup

Let's get this baby working...

```python
soup=BeautifulSoup(f, 'html.parser')
```

## Extracting all URLs

...

That's self explanatory enough

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

## Remove HTML tags and get only the text

What, what! So got rid of all the HTML tags... Yea you can still see the JavaScript stuff... For now, I am going to keep it this way!

```python
print(soup.get_text())
```
